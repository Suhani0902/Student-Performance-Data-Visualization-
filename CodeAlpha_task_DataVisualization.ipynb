```python
# ======================================================
# CodeAlpha Internship - Task 3: Data Visualization
# Project: Student Performance Data Visualization
# Author: Suhani Satapathy


# -------------------------------
# 1. Import Libraries
# -------------------------------
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn theme for clean visuals
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (8, 5)

# -------------------------------
# 2. Load Dataset

df = pd.read_csv("/content/StudentsPerformance.csv")

# Display basic info
print("Dataset Loaded Successfully!\n")
print("First 5 rows:\n", df.head(), "\n")
print("Dataset Info:\n")
print(df.info())
print("\nMissing Values:\n", df.isnull().sum())

# -------------------------------
# 3. Basic Statistics
# -------------------------------
print("\n Basic Statistical Summary:\n")
print(df.describe())

# -------------------------------
# 4. Data Cleaning & Preparation
# -------------------------------

df["average score"] = df[["math score", "reading score", "writing score"]].mean(axis=1)

# -------------------------------
# 5. Data Visualization Section
# -------------------------------
# 5.1 Gender Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x="gender", data=df, palette="Set2")
plt.title("Gender Distribution of Students")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# 5.2 Race/Ethnicity Distribution
plt.figure(figsize=(6, 4))
sns.countplot(x="race/ethnicity", data=df, palette="cool")
plt.title("Distribution of Race/Ethnicity")
plt.xlabel("Race/Ethnicity")
plt.ylabel("Count")
plt.show()

# 5.3 Test Preparation Course vs Math Score
plt.figure(figsize=(6, 4))
sns.boxplot(x="test preparation course", y="math score", data=df, palette="pastel")
plt.title("Math Score vs Test Preparation Course")
plt.xlabel("Test Preparation Course")
plt.ylabel("Math Score")
plt.show()

# 5.4 Parental Education Level vs Average Score
plt.figure(figsize=(8, 5))
sns.barplot(
    x="parental level of education",
    y="average score",
    data=df,
    palette="Set3"
)
plt.title("Average Score by Parental Level of Education")
plt.xticks(rotation=30)
plt.ylabel("Average Score")
plt.show()

# 5.5 Correlation Heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(df[["math score", "reading score", "writing score", "average score"]].corr(),
            annot=True, cmap="coolwarm", linewidths=0.5)
plt.title("Correlation Between Scores")
plt.show()

# 5.6 Gender-wise Average Scores
gender_avg = df.groupby("gender")[["math score", "reading score", "writing score"]].mean()
gender_avg.plot(kind="bar", figsize=(7, 5))
plt.title("Average Scores by Gender")
plt.ylabel("Average Score")
plt.xticks(rotation=0)
plt.show()

# 5.7 Reading vs Writing Score Scatterplot
plt.figure(figsize=(6, 4))
sns.scatterplot(x="reading score", y="writing score", hue="gender", data=df, palette="husl")
plt.title("Reading vs Writing Score (Gender-wise)")
plt.show()

# -------------------------------
# 6. Key Insights
# -------------------------------
print("\n Key Insights from Visualization:")
print("1️ Female students tend to perform better in Reading and Writing compared to males.")
print("2️ Students who completed test preparation courses scored higher in Math.")
print("3️ Higher parental education level correlates with better average student performance.")
print("4️ Reading and Writing scores have a strong positive correlation (> 0.9).")
print("5️ Race/Ethnicity group E has the highest number of students in this dataset.")

# -------------------------------
# 7. Save Visualizations (Optional)
# -------------------------------
# Uncomment if you want to save charts
# plt.savefig("output_chart.png")

print("\nData Visualization Project Completed Successfully!")
```
